```python
# PORTFOLIO OPTIMIZATION TOOL (SORTINO RATIO)

import os
import numpy as np
import pandas as pd
import yfinance as yf
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# INPUTS
tickers = ["SPY", "BND", "GLD", "QQQ", "VTI"]
start_date = "2015-01-01"
end_date = "2025-10-23"

trading_days = 252
rf = 0.04  # risk-free rate

# WEIGHTS
min_weight = 0.05
max_weight = 0.4

results_dir = "results"
os.makedirs(results_dir, exist_ok=True)
# ------------------------------------------------

# RAW DATA
prices = yf.download(tickers, start=start_date, end=end_date, auto_adjust=True)["Close"]

if isinstance(prices, pd.Series):
    prices = prices.to_frame(tickers[0])
prices = prices.sort_index().dropna(how="all")
prices = prices.dropna(axis=1, how="all")
prices = prices.ffill().dropna()

# RETURNS
returns = prices.pct_change().dropna()
log_returns = np.log(prices / prices.shift(1)).dropna()

# EXPECTED RETURN (ANNUAL)
mu = np.exp(log_returns.mean() * trading_days) - 1

# DOWNSIDE DEVIATION FUNCTION
def downside_deviation(returns, threshold=0.0):
    downside = returns[returns < threshold]
    return np.sqrt(np.mean(downside ** 2)) * np.sqrt(trading_days)

# PERFORMANCE METRICS
def portfolio_performance(weights, mu, returns, rf):
    w = np.array(weights)
    port_return = float(np.dot(w, mu.values))
    port_daily = np.dot(returns, w)
    downside_dev = downside_deviation(port_daily)
    sortino = (port_return - rf) / downside_dev if downside_dev > 0 else 0.0
    return port_return, downside_dev, sortino

def neg_sortino(weights, mu, returns, rf):
    return -portfolio_performance(weights, mu, returns, rf)[2]

# BOUNDS AND CONSTRAINTS
n = len(prices.columns)
bounds = tuple((min_weight, max_weight) for _ in range(n))
cons = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1})
x0 = np.array([1.0 / n] * n)

# SORTINO OPTIMIZATION
opt = minimize(neg_sortino, x0=x0, args=(mu, returns, rf),
               method="SLSQP", bounds=bounds, constraints=cons,
               options={'maxiter': 1000, 'ftol': 1e-9})

if not opt.success:
    raise RuntimeError(f"Optimization failed: {opt.message}")

opt_weights = pd.Series(opt.x, index=prices.columns)
opt_return, opt_dd, opt_sortino = portfolio_performance(opt.x, mu, returns, rf)

# RESULTS
out_summary = pd.Series({
    "Expected Annual Return": opt_return,
    "Downside Deviation (annualized)": opt_dd,
    "Sortino Ratio": opt_sortino,
    "Risk-free rate": rf
})
opt_weights.to_csv(os.path.join(results_dir, "optimal_weights.csv"))
out_summary.to_csv(os.path.join(results_dir, "portfolio_summary.csv"))
mu.to_csv(os.path.join(results_dir, "annual_expected_returns.csv"))
prices.to_csv(os.path.join(results_dir, "prices.csv"))

# PRINT SUMMARY
pd.set_option('display.float_format', lambda x: f"{x:.6f}")
print("\nOptimal weights:")
print(opt_weights.sort_values(ascending=False))
print("\nPortfolio summary:")
print(out_summary)

# PIE CHART
labels = [f"{t} ({w*100:.1f}%)" for t, w in opt_weights.items()]
plt.figure(figsize=(10,10))
plt.pie(opt_weights.values, labels=labels, startangle=90, counterclock=False)
plt.title(f"Sortino Ratio={opt_sortino:.3f}")
plt.tight_layout()
plt.savefig(os.path.join(results_dir, "allocation_pie.png"), dpi=300)
plt.show()
```
